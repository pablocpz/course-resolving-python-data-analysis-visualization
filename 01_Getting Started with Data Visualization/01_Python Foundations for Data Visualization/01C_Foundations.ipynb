### Groceries

```python
groceries = ["Milk", "Eggs", "Meat"]
```

### Price

```python
prices = [3.50, 2.00, 7.50]
```

## Import matplotlib library

> **Mat**hematical **Plot**ting **Lib**rary

Matplotlib is a Python library that contains functions to create Data Visualizations

```python
import matplotlib.pyplot as plt
```

### Scatter plot (points)

```python
plt.scatter(x=groceries, y=prices,
           color=["gray", "orange", "red"],
           s=[50, 100, 200],
            
           # edgecolors="black",
           # linewidths=[0.5,0.5, 0.5] ,
           marker="X");
```

### Bar plot

```python
plt.bar(x=groceries, height=prices,
           color=["gray", "orange", "red"]
       );
```

## Customize the plot

### Change the size

```python
plt.scatter(x=groceries, y=prices,
           color=["gray", "orange", "red"],
           s=[50, 100, 500],
            
           # edgecolors="black",
           # linewidths=[0.5,0.5, 0.5] ,
           marker="*");
```

### Change the color

```python
import random

#random colors [for doing manually, specify color=["green", "blue", "green"..etc]]
colors = []
for i in range(3):
    colors.append((random.random(), random.random(), random.random())) # Genera un color aleatorio para cada punto

plt.scatter(x=groceries, y=prices,
           color=colors,
           s=[50, 100, 500],
            
           # edgecolors="black",
           # linewidths=[0.5,0.5, 0.5] ,
           marker="*");
```

### Change the marker

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(x=groceries, y=prices,
           color=["gray", "orange", "red"],
           s=[50, 100, 500],
            
           # edgecolors="black",
           # linewidths=[0.5,0.5, 0.5] ,
           marker="4");
```

### Change the border

```python
plt.scatter(x=groceries, y=prices,
           color=["gray", "orange", "red"],
           s=[50, 100, 500],
            
           edgecolors="black",
           linewidths=[0.5,0.5, 0.5] ,
           marker="P");
```
