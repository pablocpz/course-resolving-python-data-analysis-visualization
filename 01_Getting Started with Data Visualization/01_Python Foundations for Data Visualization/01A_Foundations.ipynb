# Python Foundations for Data Visualization

## Create objects with data

### Football Teams

```python
football_teams = ['Real Madrid', 'Barcelona', 'Bayern Munich']
```

### Champions Leagues

```python
champions_leagues = [14, 5, 6]
```

## Import matplotlib library

> **Mat**hematical **Plot**ting **Lib**rary

Matplotlib is a Python library that contains functions to create Data Visualizations

```python
import matplotlib.pyplot as plt
```

### Scatter plot (points)

```python
plt.scatter(x=football_teams, y=champions_leagues,
           
           color=["red", "green", "blue"]);
```

### Bar plot

```python
plt.bar(x=football_teams,
        height=champions_leagues, color=["red", "green", "blue"]);
```

### Many others

https://matplotlib.org/stable/plot_types/index

```python
plt.step(football_teams, champions_leagues,
        color="purple");
```

```python
plt.stem(football_teams, champions_leagues,
        linefmt="orange");
```

```python
plt.fill_between(football_teams, champions_leagues, 
                color="pink");
```

## Customize the plot

### Change the size

```python
plt.scatter(x=football_teams, y=champions_leagues,
           
           color=["red", "green", "blue"],
           s=[400,300,100]);
```

### Change the color

Global Points

```python
plt.scatter(x=football_teams, y=champions_leagues,
           
           c=["brown"]);
```

Specific Points

```python
import random

#random colors [for doing manually, specify color=["green", "blue", "green"..etc]]
colors = []
for i in range(3):
    colors.append((random.random(), random.random(), random.random())) # Genera un color aleatorio para cada punto

plt.scatter(x=football_teams, y=champions_leagues,
           
           color=colors,
           s=[400,300,100]);
```

### Change the marker

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(x=football_teams, y=champions_leagues,
           s=[400,300,100],
           color=["red", "green", "blue"],
           marker="*");
```

```python
plt.scatter(x=football_teams, y=champions_leagues,
           s=[400,300,100],
           color=["red", "green", "blue"],
           marker="X");
```

```python
plt.scatter(x=football_teams, y=champions_leagues,
           s=[400,300,100],
           color=["red", "green", "blue"],
           marker="4");
```

### Change the border

```python
plt.scatter(x=football_teams, y=champions_leagues,
           s=[400,300,100],
           color=["red", "green", "blue"],
           marker="X",
            
        edgecolors="black", 
        linewidth=[3,2.5,1.5] #additional parameter for border size
           
   );
```
